```python
import pandas as pd
```

```python
police_df = pd.read_csv("police_ .csv")
police_df
```

```python
police_df.shape
```

```python
police_df.dtypes
```

### Convert stop_date and stop_time columns from object to date_time


```python
police_df['stop_time'] = pd.to_datetime(police_df.stop_time)
police_df['stop_time']
```

```python
police_df['stop_date'] = pd.to_datetime(police_df.stop_date)
police_df['stop_date']
```

```python
police_df.dtypes
```

```python
police_df.columns
```

### Search conducted for each male and female

```python
# Basically grouping data based on 'Gender'
pd.pivot_table(police_df,index=['driver_gender','violation'],values=['driver_age','driver_age_raw','drugs_related_stop','search_conducted'])  
```

```python
# min,max,average age of male and female for each violation and how many cases for each age on the basis of violation
```

```python
pd.pivot_table(police_df,index=['driver_gender','violation'],values= ['driver_age'],aggfunc = "mean")
```

```python
pd.pivot_table(police_df,index=['driver_gender','violation'],values= ['driver_age'],aggfunc=[max,min,len])
```

### How many people were arrested on each date

```python
police_df.groupby('stop_date').is_arrested.value_counts()  # want to find the total unique records with their counts
                                                           # apply value_count to find no.of occurences
```

### On which time police is more active

```python
police_df['stop_time'].mean()        # based on how many stops they made and exactly the time for that
```

```python
# How many males and female drivers are there in data?
police_df['driver_gender'].value_counts()        # it will give the total no.of occurences for male and female
```

```python
# Find average min,max age for male and female drivers
police_df.groupby('driver_gender').driver_age.agg(['min','max','mean'])
```

```python
police_df.tail()
```

### Check drugs activity for both male and female

```python
# basically want to check how many drugs related stops are there based on male and female.
drug_data = pd.DataFrame(police_df.groupby('driver_gender').drugs_related_stop.value_counts())
drug_data
```

```python
# Extract True Drug cases for both male and female
pd.DataFrame(police_df.groupby('driver_gender').drugs_related_stop.sum())
```

```python
# Total cases for True and False for drugs_related_stop
pd.DataFrame(police_df['drugs_related_stop'].value_counts())
```

### Action taken by police on each date

```python
# Want to check what type of action taken by police on each date
pd.DataFrame(police_df.groupby('stop_date').stop_outcome.value_counts())
```

### Warning cases for male and female

```python
# want to check how many times police has given a warning as stop_outcome
warning_data = police_df[police_df['stop_outcome'] == 'Warning']
warning_data
```

```python
len(police_df[police_df['stop_outcome'] == 'Warning'])  # get the total no.of values
```

```python
# check warning cases based on male and female
pd.DataFrame(warning_data.groupby('driver_gender').stop_outcome.value_counts())
```

```python
# How many male and female for each violation
pd.crosstab(police_df['driver_gender'],police_df['violation'])      # want to comapre two and more columns
```

```python
pd.crosstab(police_df['driver_gender'],police_df['violation'],margins=True)
```
